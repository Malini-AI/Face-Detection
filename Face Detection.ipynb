```python
import cv2
import PIL.Image
```

```python
# Load the cascade
face_cascade=cv2.CascadeClassifier('C:/Users/malin/Desktop/haarcascade_frontalface_default.xml')
face_cascade
```

```python
# Read the input image
img = cv2.imread('C:/Users/malin/Desktop/bts.jpg')
```

```python
# Convert into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# Detect faces
faces = face_cascade.detectMultiScale(gray, 1.1, 4)
```

```python
# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
# Display the output
cv2.imwrite('C:/Users/malin/Desktop/face_deteted.jpg', img)
```
